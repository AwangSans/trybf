```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# 
processed_data = np.load('final1.npy')
array_2d_1 = np.array(processed_data).reshape(24, -1)
processed_data = np.load('final2.npy')
array_2d_2 = np.array(processed_data).reshape(24, -1)

array_2d = np.hstack((array_2d_1, array_2d_2))[:, :-2]

# 
plt.imshow(array_2d, cmap='hsv', aspect='auto')
plt.colorbar(label='Value')
plt.title('Time - Frequency Diagram Start from Aug 25th')

y_ticks = np.arange(len(array_2d))  # 
y_labels = [i * 0.02 for i in range(1, 25)]

start = 7 * 8
over = 15 * 8 - 4
plt.axvline(x=start, linestyle='--', color='r', label='start')
plt.axvline(x=over, linestyle='--', color='r')

plt.ylabel('Frequency(Hz)')
plt.xlabel('Time')
plt.yticks(ticks=y_ticks, labels=y_labels)

# 
start_date = datetime(2025, 8, 25)
x_ticks = np.arange(0, 176, 30)
x_dates = [start_date + timedelta(days=x / 8) for x in x_ticks]

# 
x_date_labels = [date.strftime('%m-%d') for date in x_dates]

#  x 
plt.xticks(x_ticks, x_date_labels)

#  start  over 
new_x_dates = [start_date + timedelta(days=start / 8), start_date + timedelta(days=over / 8)]
new_x_date_labels = [date.strftime('%m-%d') for date in new_x_dates]
#new_xticks = [start, over]
all_xticks = list(plt.xticks()[0]) #+ new_xticks
all_x_date_labels = list(plt.xticks()[1]) #+ new_x_date_labels
plt.xticks(all_xticks, all_x_date_labels)

#  x 
plt.xticks(rotation=0)

#  y  0.02 
plt.gca().invert_yaxis()

plt.show()
    
```
